Libraries

```python
import pandas as pd
import numpy as np
```

Import data

```python
rookies_2021 = pd.read_csv('2021nbarookie.csv', header = [1])
rookies_2020 = pd.read_csv('2020nbarookie.csv', header = [1])
rookies_2019 = pd.read_csv('2019nbarookie.csv', header = [1])
rookies_2018 = pd.read_csv('2018nbarookie.csv', header = [1])
rookies_2017 = pd.read_csv('2017nbarookie.csv', header = [1])
rookies_2016 = pd.read_csv('2016nbarookie.csv', header = [1])
rookies_2015 = pd.read_csv('2015nbarookie.csv', header = [1])
rookies_2014 = pd.read_csv('2014nbarookie.csv', header = [1])

odds_2021 = pd.read_csv('2021nbaodds.csv')
odds_2020 = pd.read_csv('2020nbaodds.csv')
odds_2019 = pd.read_csv('2019nbaodds.csv')
odds_2018 = pd.read_csv('2018nbaodds.csv')
odds_2017 = pd.read_csv('2017nbaodds.csv')
odds_2016 = pd.read_csv('2016nbaodds.csv')
odds_2015 = pd.read_csv('2015nbaodds.csv')
odds_2014 = pd.read_csv('2014nbaodds.csv')

abbrev = pd.read_csv('abbreviations.csv')
```

Cleaning!

```python
#adding column for year
rookies_2021['year'] = 2021
rookies_2020['year'] = 2020
rookies_2019['year'] = 2019
rookies_2018['year'] = 2018
rookies_2017['year'] = 2017
rookies_2016['year'] = 2016
rookies_2015['year'] = 2015
rookies_2014['year'] = 2014

```

```python
odds_2021['year'] = 2021
odds_2020['year'] = 2020
odds_2019['year'] = 2019
odds_2018['year'] = 2018
odds_2017['year'] = 2017
odds_2016['year'] = 2016
odds_2015['year'] = 2015
odds_2014['year'] = 2014
```

```python
#appending stuff 
rookies_8yrs = pd.concat([rookies_2021, rookies_2020, rookies_2019, rookies_2018, rookies_2017, rookies_2016, 
                          rookies_2015, rookies_2014])

odds_8yrs = pd.concat([odds_2021, odds_2020, odds_2019, odds_2018, odds_2017, odds_2016, 
                       odds_2015, odds_2014])
```

```python
#team 
rookies_8yrs['Team'] = rookies_8yrs['Debut'].astype(str).str.split().str[3]

#name
rookies_8yrs[['Player', 'ID']] = rookies_8yrs['Player'].str.split(pat = "\\", expand=True)
```

```python
#result
odds_8yrs['win'] = odds_8yrs['Result'].astype(str).str.split('-').str[0].astype(int)
```

```python
#need to do abbreviations to match up data
odds_8yrs = pd.merge(odds_8yrs, abbrev)

#checking merge of team abbreviations
odds_8yrs.value_counts(["Team", "Abbrev"])
```

```python
#create new vars to do the merge
odds_8yrs['year_team'] = odds_8yrs['Abbrev'] + '_' + odds_8yrs['year'].astype(str)
rookies_8yrs['year_team'] = rookies_8yrs['Team'] + '_' + rookies_8yrs['year'].astype(str)
```

```python
#Merge data

merged_8yrs = pd.merge(rookies_8yrs,odds_8yrs,on='year_team')
```

```python
#great. Let's drop some USELESS stuff
print(merged_8yrs.columns)
merged_8yrs.drop(['Rk', 'Result', 'Abbrev', 'year_x', 'Unnamed: 2', 'Odds', 'ID', '-9999', 'Team_x',
                  'Team_y'], axis=1, inplace=True)
print(merged_8yrs.columns)
```

```python
#create winner variable
# merged_8yrs['winner'] = ['YES' if a > 41 else 'NO' for a in merged_8yrs['win']]

def conditions(r):
    if (r['year_y'] in [2019, 2020]) and (r['win'] > 36):
        return 'YES'
    elif (r['win']>41):
        return 'YES'
    else:
        return 'NO'
    
merged_8yrs['winner'] = merged_8yrs.apply(conditions, axis=1)

merged_8yrs.head(50)
```

```python
#have to subset dataset 

def conditions(s):
    if (s['year_y'] in [2019, 2020]) and (s['G'] > 25):
        return 1
    elif (s['G']>29):
        return 1
    else:
        return 0 
    
merged_8yrs['contrib'] = merged_8yrs.apply(conditions, axis=1)

merged_8yrs.head(50)

```

```python
#drop non contributors
merged_8yrs = merged_8yrs[merged_8yrs.contrib ==1 ]
print(merged_8yrs.Age.mean())
merged_8yrs.head(50)

```

Export

```python
merged_8yrs.to_csv(r'final.csv', index = False)
```
